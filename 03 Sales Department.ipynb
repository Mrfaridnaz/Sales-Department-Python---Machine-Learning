```python
# StateHoliday: indicates a state holiday. Normally all stores, with few exceptions, are closed on state holidays. Note that all schools are closed on public holidays and weekends. a = public holiday, b = Easter holiday, c = Christmas, 0 = None
# SchoolHoliday: indicates if the (Store, Date) was affected by the closure of public schools
```

```python
from prophet import Prophet

def sales_prediction(Store_ID, sales_df, holidays, periods):
    # Function that takes in the storeID, sales_df, holidays, and number of future periods to forecast.
    # It returns the forecasted sales data and plots
    
    # Filter the dataframe by the specified Store_ID
    sales_df = sales_df[sales_df['Store'] == Store_ID]
    
    # Rename columns to fit Prophet's required format
    sales_df = sales_df[['Date', 'Sales']].rename(columns={'Date': 'ds', 'Sales': 'y'})
    
    # Sort values by the 'ds' column (date)
    sales_df = sales_df.sort_values('ds')
    
    # Initialize Prophet model, including holidays
    model = Prophet(holidays=holidays)
    
    # Fit the model with the sales data
    model.fit(sales_df)
    
    # Create a future dataframe for the specified number of periods
    future = model.make_future_dataframe(periods=periods)
    
    # Make predictions using the future dataframe
    forecast = model.predict(future)
    
    # Plot the forecast
    figure = model.plot(forecast, xlabel='Date', ylabel='Sales')
    
    # Plot the forecast components (trend, weekly, yearly, holidays)
    figure2 = model.plot_components(forecast)
    
    return forecast, figure, figure2

# Call the function with an example Store_ID, sales dataframe, holidays dataframe, and number of future periods
# sales_prediction(10, sales_train_all_df, holidays_df, 60)
```
